```python
import nbodykit.lab as nlab
```

```python
print(nlab.cosmology)
```

```python
cosmo = nlab.cosmology.Planck15
```

```python
print(cosmo)
```

```python
Plin = nlab.cosmology.EHPower(cosmo, redshift=0.55)
```

```python
help(nlab.cosmology.EHPower)
```

```python
help(nlab.LogNormalCatalog)
```

```python
help(nlab.ArrayCatalog)
```

```python
help(nlab.ArrayCatalog)
```

```python
source = nlab.LogNormalCatalog(Plin=Plin, nbar=3.e-7, BoxSize=1380., Nmesh=8, seed=42)
```

```python
print(source.attrs)
```

```python
print(source)
```

```python
source['Position'] += source['VelocityOffset']*[0, 0, 1]
```

```python
# compute P(k,mu) and multipoles
result = nlab.FFTPower(source, mode='2d', poles=[0,2,4], los=[0,0,1])
```

```python
print(result.power)
```

```python
# and save
output = "./nbkit_example_power.json"
result.save(output)
```
